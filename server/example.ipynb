```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import HDBSCAN
import pandas as pd
```

```python
url = "C:\\Users\\Abhi\\Desktop\\python_projects\\Accident Data Analysis-20240312T073202Z-001\\Accident Data Analysis\\Copy_of_AccidentReports.csv"
data = pd.read_csv(url)
data = data[['Latitude', 'Longitude']]
data = data[(data['Latitude']!=0.0) & (data['Longitude']!=0.0)]
```

```python
data = data[(data['Latitude']<=14) & (data['Latitude']>=13) & (data['Longitude']<=78)& (data['Longitude']>=76)]
print(data.describe())
```

```python
X = data.to_numpy()
model = HDBSCAN(min_cluster_size=10).fit(X)
```

```python
clusters = model.labels_
```

```python
print(clusters)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=clusters)
plt.show()
```

```python
data['color'] = clusters
```
